### 2. Stratified Sampling

```python
cust = pd.read_csv("./customer_shopping_data.csv")
```

```python
cust
```

```python
# Define the criteria for stratification (e.g. age groups)
age_groups = pd.cut(cust['age'], bins=[18, 30, 45, 60, 100], labels=['Adults', 'Mid-aged', 'Older', '60+'])

# Perform stratified sampling
stratified_sample = cust.groupby(age_groups).apply(lambda x: x.sample(frac=0.2)) # Select 30% samples from each age group
```

```python
stratified_sample
```

### 3. Systematic Sampling

```python
# let us define the sampling interval 
n = 6

# perform systematic sampling
systematic_sample = cust.iloc[::n, :]
# it selects every nth row from the original data
```

```python
systematic_sample
```

### 6. Custom aggregations

```python
# sum, mean, min, max
-------------
custom aggregations,
we will be definging our own functions based on our requirements
# custom defined functions
```

```python
import numpy as np
```

```python
players
```

```python
# define a custom aggregation function
def custom_agg(x):
    return np.sum(x) / np.count_nonzero(x)
```

```python
# apply custom aggregation function on a column
custom_aggregated_value = players['Height(cm)'].agg(custom_agg)
```
